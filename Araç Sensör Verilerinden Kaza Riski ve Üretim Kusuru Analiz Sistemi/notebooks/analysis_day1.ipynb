```python
import pandas as pd

df = pd.read_csv(r"C:\Users\aayse\OneDrive\Masaüstü\Araç Sensör Verilerinden Kaza Riski ve Üretim Kusuru Analiz Sistemi\data\car_data.csv.csv")

# Veri çerçevesinin ilk 5 satırını gösterir.
df.head()
```

```python
# Veri setinin satır ve sütun sayısını gösterir
print(df.shape) 

# Her sütunun veri tipini ve eksik değer olup olmadığını gösterir
df.info()
```

```python
df.isnull().sum()
```

```python
#Boş yerleri az miktarda eksik olduğu için en sık görülen değerlerle doldur.
df['Engine Fuel Type']=df['Engine Fuel Type'].fillna(df['Engine Fuel Type'].mode()[0])
df['Number of Doors'] = df['Number of Doors'].fillna(df['Number of Doors'].mode()[0])

#Boş yerleri orta mikatarda eksik olduğu için median ile doldur.
df['Engine HP']= df['Engine HP'].fillna(df['Engine HP'].median())
df['Engine Cylinders'] = df['Engine Cylinders'].fillna(df['Engine Cylinders'].median())

#Boş yerleri çok fazla eksik olduğu için 'unknown' ile doldur.
df['Market Category']=df['Market Category'].fillna("unknown")
```
